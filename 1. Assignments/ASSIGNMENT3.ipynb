```python
IQR = df['SepalWidthCm'].quantile(0.75)-df['SepalWidthCm'].quantile(0.25)
IQR
```

```python
upperBound=df['SepalWidthCm'].quantile(0.75)+(1.5*IQR)
upperBound
```

```python
df['SepalWidthCm']=np.where(df['SepalWidthCm']>upperBound,upperBound,df['SepalWidthCm'])
df['SepalWidthCm']=np.where(df['SepalWidthCm']<lowerBound,lowerBound,df['SepalWidthCm'])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['Species'] = le.fit_transform(df['Species'])
```

```python
df['Species']
```

```python
x = df.drop(columns=['Species'],axis = 1)
y = df['Species']
```

```python
y
```

# Splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=210)
```

# Model Training

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn import metrics
r2Score = metrics.confusion_matrix(y_test,y_pred)
r2Score
```
